# Pie chart

```python
brands = ['onePlus','Apple','Samsung','Nokia','Realme']
popularity = [22,34,30,3,10]
c = ['red','silver','cyan','pink','orange']

plt.pie(popularity, labels=brands, colors = c)
plt.show()
```

```python
ex = [0,0,0,0,0.1]

plt.pie(popularity, labels=brands, colors = c, explode=ex)
plt.show()
```

```python
# plt.pie(popularity, labels=brands, colors = c, explode=ex,autopct="%.f")
plt.pie(popularity, labels=brands, colors = c, explode=ex,autopct="%.2f")
plt.show()
```

```python
# plt.pie(popularity, labels=brands, colors = c, explode=ex,autopct="%.2f",startangle=90)
plt.pie(popularity, labels=brands, colors = c, explode=ex,autopct="%.2f",startangle=90)
plt.savefig("telecom.png", facecolor='black')
plt.show()
```

# Stack Plot

```python
days = [1,2,3,4,5,6,7]

NOP1 = [5,10,30,20,35,60,80]
NOP2 = [10,20,30,25,40,50,90]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days, NOP1, NOP2, NOP3, colors=['red','black','grey'], labels = ['week1','week2','week3'])
plt.legend()
plt.show()
```
