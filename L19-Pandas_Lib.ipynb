# Merge, Join and Concatenate

```python
data1 = {
    "Emp Id": ["E01", "E02", "E03", "E04", "E05", "E06"],
    "Names":  ["Ram", "Shyam", "Rahul", "Vishal", "Ravi", "John"],
    "Age":    [34, 56, 23, 44, 32, 36]
}

data2 = {
    "Emp Id": ["E01", "E02", "E03", "E04", "E05", "E06"],
    "Salary": [45000, 56000, 34000, 30000, 50000, 48000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print(df1)
print()
print(df2)
```

```python
df3 = pd.merge(df1,df2, on="Emp Id")
df3
```

This dataset dont have any null or change values if it have the result will be different

```python
data1 = {
    "Emp Id": ["E01", "E02", "E03", "E04", "E05", "E06"],
    "Names":  ["Ram", "Shyam", "Rahul", "Vishal", "Ravi", "John"],
    "Age":    [34, 56, 23, 44, 32, 36]
}

data2 = {
    "Emp Id": ["E01", "E07", "E03", "E04", "E05", "E08"],
    "Salary": [45000, 56000, 34000, 30000, 50000, 48000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df3 = pd.merge(df1,df2,on='Emp Id')
df3
```

```python
df3 = pd.merge(left = df1,right = df2,on ='Emp Id',how ='left')
df3 # These datasets are now joining based on df1 (left dataset)
```

```python
df3 = pd.merge(left = df1,right = df2,on ='Emp Id',how ='right')
df3 # These datasets are now joining based on df2 (right dataset)
```

```python
data1 = {
    "Emp Id": ["E01", "E02", "E03", "E04", "E05", "E06"],
    "Names":  ["Ram", "Shyam", "Rahul", "Vishal", "Ravi", "John"],
    "Age":    [34, 56, 23, 44, 32, 36]
}

data2 = {
    "Emp Id": ["E07", "E08", "E09", "E010", "E011", "E012"],
    "Names":  ["krishna", "karan", "ali", "dabbu", "bunty", "golu"],
    "Age":    [28, 23, 22, 19, 18, 13]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df3 = pd.concat([df1, df2])
df3
```

# Pivot and Melting

```python
dict1 = {
    "keys":["k1","k2","k1","k2"],
    "Name": ["john","ben","david","peter"],
    "Houses":["red","blue","green","red"],
    "Grades":["3rd","8th","9th","8th"]
}

df = pd.DataFrame(dict1)
df
```

```python
df.pivot(index="keys", columns="Name", values="Houses")
```

```python
df.pivot(index="keys", columns="Name", values=["Houses","Grades"])
```

```python
dict2 = {
    "Name": ["john","ben","lisa","peter"],
    "Gender":["m","m","f","m"],
    "Age":["28","31","30","27"]
}
df = pd.DataFrame(dict2)
```

```python
pd.melt(df, id_vars=['Name'],value_vars=["Gender",'Age'])
```

```python
pd.melt(df, id_vars=['Name'],value_vars=["Gender",'Age'], var_name="Gender&Age")
```

```python
pd.melt(df, id_vars=['Name'],value_vars=["Gender",'Age'], var_name="Gender&Age", value_name="Data")
```
