# Line Plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "Days":[1,2,3,4,5,6,7],
    "Nop":[50,40,50,20,10,40,30]
}

df = pd.DataFrame(data)
df.head()
```

```python
sns.lineplot(data = df, x="Days", y="Nop")
plt.show()
```

```python
data = pd.read_excel("ESD.xlsx")
data.head(2)
```

```python
sns.lineplot(data=data, x="Business Unit", y="Age", hue="Gender")
plt.show()
```

```python
sns.lineplot(data=data, x="Business Unit", y="Age", hue="Ethnicity", style="Gender")
plt.show()
```

```python
sns.lineplot(data=data, x="Business Unit", y="Age",style="Gender")
plt.show()
```

```python
sns.lineplot(data=data, x="Business Unit", y="Age",style="Ethnicity")
plt.show()
```

```python
color = sns.color_palette("viridis")
# color = sns.color_palette("GnBu")
sns.lineplot(data=data, x="Business Unit", y="Age", hue="Ethnicity", palette = color)
plt.show()
```

# Bar Plot

```python
data = sns.load_dataset("tips")
# data = sns.load_dataset("titanic")
data.head()
```

```python
sns.barplot(data=data, x="day",y="tip", estimator="median")
# sns.barplot(data=data, x="day",y="tip", estimator="mean")
# sns.barplot(data=data, x="day",y="tip", estimator="sum")
plt.show()
```

```python
sns.barplot(data=data, x="day",y="tip",hue="sex", palette="viridis")
plt.show()
```

```python
sns.barplot(data=data, x="day",y="tip",hue="sex", palette="viridis", order=['Sun','Sat','Fri','Thur'])
plt.show()
```

```python
sns.barplot(data=data, x="day",y="tip",hue="sex", palette="viridis",errorbar=('ci',0))
plt.show()
```

# Scatter Plot

# Heatmap

```python
gp = data.groupby('day').agg({'tip':'mean'})
sns.heatmap(data=gp)
plt.show()
```

```python
data = pd.read_excel("ESD.xlsx")
gp = data.groupby("Job Title").agg({"Annual Salary":"mean"})
sns.heatmap(gp)
plt.show()
```
